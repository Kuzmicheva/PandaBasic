Задание 1
Скачайте с сайта http://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.



```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

```python
movies_5_0 = joined[joined.rating == 5.0]
movies_5_0.head()
```

```python
answer = movies_5_0['title'].value_counts().head(1)
```

```python
answer = answer.to_string()
```

```python
movie = answer[0:32]
```

```python
number_of_ratings = answer[36:]
```

```python
print(f"Фильму {movie},  поставили больше всего оценок 5.0: {number_of_ratings}")
```

Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.



```python
data = pd.read_csv('power.csv')
data.head()
```

```python
filtered_countries = data[ (data['country']=='Lithuania') | (data['country']=='Latvia') | (data['country']=='Estonia') ]

filtered_countries.head()
```

```python
filtered_countries_category_4_12_21 = filtered_countries[ (filtered_countries['category']== 4) | (filtered_countries['country']== 12) | (filtered_countries['country']== 21) ]
```

```python
filtered_countries_category_4_12_21
```

```python
filtered_countries_category_4_12_21_no_negative = filtered_countries[ (filtered_countries['quantity'] >= 0) ]
```

```python
filtered_countries_category_4_12_21_no_negative.sort_values(by='quantity').head()
```

```python
filtered_countries_category_4_12_21_no_negative.query("2005 <= year < 2010").head()
```
